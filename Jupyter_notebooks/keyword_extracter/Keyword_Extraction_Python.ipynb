```python
import nltk

from nltk.stem.porter import PorterStemmer

path = './tf-idf'
token_dict = {}


def tokenize(text):
    tokens = nltk.word_tokenize(text)
    stems = []
    stems2= []
    for item in tokens:
        stems.append(PorterStemmer().stem(item))
        stems2.append(singularize(str(item)))
    return stems, stems2
```

```python
def clean_text(document):
    return remove_stop_words(remove_punctuation(document).lower())

def remove_punctuation(document):
    # initializing punctuations string
    punc = '''![]{};:'"\,<>./?@#$%^&*_~'''

    # Removing punctuations in string
    # Using loop + punctuation string
    for ele in document:
        if ele in punc:
            document = document.replace(ele, "")
    return document

# Remove stopwords
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
def remove_stop_words(document):
    stop_words=stopwords.words('english')
    all_words = document.split()
    for each_word in all_words:
        if each_word  in stop_words:
            document = document.replace(each_word + ' ', '')
    return document

clean_text(doc3)
```

```python
# TfidfVectorizer 
# CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
import pandas as pd
# set of documents
train = ['The sky is blue Molecule.','The molecule sun is bright. molecules. works. drink and food']
test = ['The sun in the sky is bright', 'We can see the shining sun, the bright sun.']
# train =[doc,doc1, doc2, doc3]
# instantiate the vectorizer object
voc = ['sky', 'sun']
countvectorizer = CountVectorizer(vocabulary= voc,analyzer= 'word', stop_words='english')
tfidfvectorizer = TfidfVectorizer(strip_accents='ascii', analyzer='word',stop_words= 'english', ngram_range = (1, 2))
# convert the documents into a matrix
count_wm = countvectorizer.fit_transform(train)
tfidf_wm = tfidfvectorizer.fit_transform(train)
#retrieve the terms found in the corpora
# if we take same parameters on both Classes(CountVectorizer and TfidfVectorizer) , it will give same output of get_feature_names() methods)
#count_tokens = tfidfvectorizer.get_feature_names() # no difference
count_tokens = countvectorizer.get_feature_names()
tfidf_tokens = tfidfvectorizer.get_feature_names()
index = ['doc','doc1', 'doc2', 'doc3']
index = ['doc','doc1']
df_countvect = pd.DataFrame(data = count_wm.toarray(),index = index,columns = count_tokens)
df_tfidfvect = pd.DataFrame(data = tfidf_wm.toarray(),index =index,columns = tfidf_tokens)
print("Count Vectorizer\n")
print(df_countvect)
print("\nTD-IDF Vectorizer\n")
print(df_tfidfvect)
```
