```python
import pandas as pd
```

```python
concordance=pd.read_excel('Concordantie_VODS_V2018_WoS.xlsx', index_col=0, skiprows=[0,1])
concordance.index=pd.Series(concordance.index).apply(lambda x: x.capitalize())
matrix15=pd.read_excel('matrix15.xlsx', index_col=0)
matrix15
```

```python
concordance
```

```python
new_matrix=[]
for sc in matrix15.index:
    new_rows=[]
    for code in concordance[concordance.index == sc]['VODS niveau 2']:
        new_rows.append([code[:4]]+list(matrix15[matrix15.index == sc].iloc[0,]/len(concordance[concordance.index == sc]['VODS niveau 2'])))
    new_matrix= new_matrix+new_rows

half_translated=pd.DataFrame(new_matrix,columns=['code']+matrix15.columns.to_list())
```

```python
half_translated=half_translated.groupby(['code']).sum()
```

```python
half_translated=half_translated.T
```

```python
new_matrix=[]
for sc in half_translated.index:
    new_rows=[]
    for code in concordance[concordance.index == sc]['VODS niveau 2']:
        new_rows.append([code[:4]]+list(half_translated[half_translated.index == sc].iloc[0,]/len(concordance[concordance.index == sc]['VODS niveau 2'])))
    new_matrix= new_matrix+new_rows
completely_translated=pd.DataFrame(new_matrix,columns=['code']+half_translated.columns.to_list())
```

```python
completely_translated=completely_translated.groupby(['code']).sum()
```

```python
completely_translated=completely_translated.T
```
