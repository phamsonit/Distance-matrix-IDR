```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from ast import literal_eval
from math import log
```

```python
# load similarity matrix M (between disciplines)

cur_path = os.getcwd()
count_type = 'disc_in_a_person'
data = 'pub_2014'

# count_type = 'colab'
# data = 'projects'
filename ='sm_ochiai_' + count_type + '_' + data+  '.txt'
new_path = os.path.relpath('..\\2_Discipline_distances/'+filename, cur_path)
ochiai = np.loadtxt(new_path)
```

```python
def lower_triangle(matrix):
    lower=[]
    for i in range(len(matrix)):
        for j in range(i):
            lower=lower+[matrix[i,j]]
    return lower
```

```python
f1= 1-ochiai**(log(2,42))
f2 =(1-ochiai)**(log(2)/log(42/41))
a=1/(42*41)

f3 =abs( np.log((a+ochiai)/(a+1)) /(np.log(a/(a+1))))
```

```python
# Fixing random state for reproducibility 
np.random.seed(19680801) 
    
# create random data 
n=100
xdata = np.array(range(0,n))/(n-1)
xdata= np.concatenate((xdata,np.random.random(200)/20))    

    
# sort the data so it makes clean curves 
xdata.sort() 

    
# create some y data points 
ydata1 =  1 - xdata**(log(2,42))
ydata2 = (1-xdata)**(log(2)/log(42/41))
a=1/(41*42)
ydata3 = np.log((a+xdata)/(a+1)) /np.log(a/(a+1))
    
plt.subplots(figsize=(12, 4))   
    
# plot the data 
plt.subplot(1, 2, 1)
plt.plot(xdata, ydata1, color ='tab:blue', label=r'$f_1$') 
plt.plot(xdata, ydata2, color ='tab:orange', label=r'$f_2$') 
plt.plot(xdata, ydata3, color ='tab:green', label=r'$f_3$')    
    
# set the limits 
plt.xlim([0, 1]) 
plt.ylim([0, 1]) 

plt.xlabel('Similarity')
plt.ylabel('Distance')

plt.legend()

# plot the data 
plt.subplot(1, 2, 2)
plt.plot(xdata, ydata1, color ='tab:blue', label=r'$f_1$') 
plt.plot(xdata, ydata2, color ='tab:orange', label=r'$f_2$') 
plt.plot(xdata, ydata3, color ='tab:green', label=r'$f_3$')   
    
# set the limits 
plt.xlim([0, 0.05]) 
plt.ylim([0, 1]) 

plt.xlabel('Similarity')
plt.ylabel('Distance')

plt.legend()

# display the plot 
plt.show()
```
