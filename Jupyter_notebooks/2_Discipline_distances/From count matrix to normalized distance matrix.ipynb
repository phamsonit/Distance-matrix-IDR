## Cosine salton normalization

```python
count_type = 'citation_flow'
data = 'matrix15'
count_type = 'disc_in_a_person'
data = 'pub_2014'
count_type = 'colab'
data = 'projects'

df = pd.read_table(count_type+'_'+data+'.txt', delim_whitespace=True, header=None)
nb_of_disciplines = df.shape[0]  # 42


cosine_sim_matrix=np.zeros((nb_of_disciplines, nb_of_disciplines))


for i in range(nb_of_disciplines):
    df.iloc[i,i]=sum(df.iloc[:,i])


for i in range(nb_of_disciplines):
    for j in range(nb_of_disciplines):
        if sum(df.iloc[:,i]) == 0 or sum(df.iloc[:,j]) == 0:
            if i == j:
                cosine_sim_matrix[i,j]=1
            else:
                cosine_sim_matrix[i,j]=0
        else:         
            cosine_sim_matrix[i,j] = np.inner(df.iloc[:,i],df.iloc[:,j])/(np.inner(df.iloc[:,j],df.iloc[:,j])**(1/2)* np.inner(df.iloc[:,i],df.iloc[:,i])**(1/2))


cosine_distance_matrix=np.round(1-cosine_sim_matrix,10)

# np.savetxt('dm_cosine_' + count_type + '_' + data+  '.txt', cosine_distance_matrix)
np.savetxt('sm_cosine1_' + count_type + '_' + data+  '.txt', 1-cosine_distance_matrix)
cosine_distance_matrix
```

## Jaccard index plus normalization

```python
count_type = 'disc_in_a_person'
data = 'pub_2014'

df = pd.read_table(count_type+'_'+data+'.txt', delim_whitespace=True, header=None)
nb_of_disciplines = df.shape[0]  # 42

jaccard1 = np.zeros((42,42))
for i in range(42):
    jaccard1[i,i] = 1
    for j in range(i):
        jaccard1[i,j]=df.iloc[i,j]/(df.iloc[i,i]+df.iloc[j,j].sum()-df.iloc[i,j])
        jaccard1[j,i]=jaccard1[i,j]
np.savetxt('dm_jaccard_' + count_type + '_' + data+  '.txt',1- jaccard1)
np.savetxt('dm_jaccard_sqrt_' + count_type + '_' + data+  '.txt',1- np.sqrt(jaccard1))
```

## Sorensen Dice Coefficient

```python
count_type = 'colab'
data = 'projects'
df = pd.read_table(count_type+'_'+data+'.txt', delim_whitespace=True, header=None)
nb_of_disciplines = df.shape[0]  # 42
df= df+1


sdc = np.zeros((42,42))
for i in range(42):
    sdc[i,i] = 1
    for j in range(i):
        sdc[i,j]=(2*df.iloc[i,j])/(df.iloc[i,:].sum()+df.iloc[:,j].sum())
        sdc[j,i]=sdc[i,j]
np.savetxt('dm_sdc_' + count_type + '_' + data+  '.txt', -np.log(sdc)/np.log(42))
np.savetxt('sm_sdc_' + count_type + '_' + data+  '.txt', sdc)
```

## Ochiai Coefficient

```python
count_type = 'colab'
data = 'projects'
df = pd.read_table(count_type+'_'+data+'.txt', delim_whitespace=True, header=None)

ochiai = np.zeros((42,42))
for i in range(42):
    ochiai[i,i] = 1
    for j in range(i):
        ochiai[i,j]=(df.iloc[i,j])/(df.iloc[i,:].sum()*df.iloc[:,j].sum())**(1/2)
        ochiai[j,i]=ochiai[i,j]

np.savetxt('sm_ochiai_' + count_type + '_' + data+  '.txt', ochiai)
```
